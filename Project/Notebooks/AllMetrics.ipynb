```python
import warnings
warnings.filterwarnings("ignore")
```

```python
summaries = pd.read_csv("summaries.csv", index_col=0)
```

```python
rouge = evaluate.load("rouge")
bleu  = evaluate.load("bleu")
bertscore  = evaluate.load("bertscore")
meteor = evaluate.load('meteor')
```

```python
def get_metrics(generated_summaries, reference_summaries):
    metrics_dict = rouge.compute(predictions=generated_summaries, references=reference_summaries)
    metrics_dict["bleu"] = bleu.compute(predictions=generated_summaries, references=reference_summaries)["bleu"]
    metrics_dict["meteor"] = meteor.compute(predictions=generated_summaries, references=reference_summaries)["meteor"]
    return metrics_dict
```

```python
columns = ["rouge1", "rouge2", "rougeL", 'rougeLsum', 'bleu', 'meteor']
indexes = ["TextRank", "LexRank", "LSA", "KL", "Luhn", "BertSum"]

result_metrics = pd.DataFrame(columns=columns, index=indexes)
```

```python
for ind in indexes:
    metrics_dict = get_metrics(summaries[ind].values, summaries["References"].values)
    for col in columns:
        result_metrics.loc[ind][col] = metrics_dict[col]
```

```python
for ind in ["LexRank", "LSA", "KL", "Luhn", "BertSum"]:
    
    metrics = bertscore.compute(predictions=summaries[ind].values, references=summaries["References"].values, lang='ru')
    
    metrics_dict = {'precision': np.mean(metrics['precision']), 
                 'recall': np.mean(metrics['recall']),
                'f1': np.mean(metrics['f1'])}
    
    bert_metrics.loc[ind]['bert_precision'] = metrics_dict['precision']
    bert_metrics.loc[ind]['bert_recall'] = metrics_dict['recall']
    bert_metrics.loc[ind]['bert_f1'] = metrics_dict['f1']
```
